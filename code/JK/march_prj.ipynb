```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('data/test_deeplearning.txt')

docs = loader.load()

text = ''
for i in docs :
    text += i.page_content

print(text)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 50,
    chunk_overlap = 10,
    separators = ['\n\n', '\n', '.', ' ']
)

chunks = text_splitter.split_text(text)

chunks
```

```python
import pandas as pd

from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA
from langchain.llms import HuggingFacePipeline

from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch
```

```python
# ì„ë² ë”© ìƒì„±
embedding_model_name = "jhgan/ko-sbert-nli"  # ì„ë² ë”© ëª¨ë¸ ì„ íƒ
embedding = HuggingFaceEmbeddings(model_name=embedding_model_name)

# ë²¡í„° ìŠ¤í† ì–´ì— ë¬¸ì„œ ì¶”ê°€
test_vector_store = FAISS.from_texts(chunks, embedding)
```

```python
retriever = test_vector_store.as_retriever()
# ê´€ë ¨ ë¬¸ì„œë¥¼ ê²€ìƒ‰
docs = retriever.invoke("yoloì— ëŒ€í•´ ì„¤ëª…í•´ì£¼ì„¸ìš”")

for doc in docs:
    print(doc.page_content)
    print("=========================================================")
```

```python
from pptx import Presentation
from pptx.util import Inches
import os
import subprocess

def convert_mp3_to_wav(mp3_path, wav_path):
    """MP3ë¥¼ PowerPoint í˜¸í™˜ì„±ì´ ë†’ì€ WAV í¬ë§·ìœ¼ë¡œ ë³€í™˜"""
    command = ["ffmpeg", "-i", mp3_path, "-acodec", "pcm_s16le", "-ar", "44100", wav_path]
    subprocess.run(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

def add_audio_to_ppt(ppt_path, num_slides):
    """ì›ë³¸ PPTì— ìŠ¬ë¼ì´ë“œë³„ ë°œí‘œ ìŒì„±ì„ ìë™ ì‚½ì…í•˜ê³ , ì „í™˜ ì‹œ ìë™ ì¬ìƒ ì„¤ì •"""
    prs = Presentation(ppt_path)

    for i in range(num_slides):
        if i < len(prs.slides):
            slide = prs.slides[i]
            mp3_path = f"speech_slide_{i+1}.mp3"
            wav_path = '../../data/1000.wav'

            if os.path.exists(wav_path):
                # MP3 â†’ WAV ë³€í™˜
                # convert_mp3_to_wav(mp3_path, wav_path)

                # ìŠ¬ë¼ì´ë“œì— ì˜¤ë””ì˜¤ ì‚½ì…
                shape = slide.shapes.add_shape(1, Inches(1), Inches(1), Inches(1), Inches(1))
                shape.text = "ë°œí‘œ ìŒì„±"

                slide.shapes.add_movie(wav_path, left=Inches(1), top=Inches(1), width=Inches(1), height=Inches(1))

                print(f"Slide {i+1}ì— ì˜¤ë””ì˜¤ ì¶”ê°€ ì™„ë£Œ: {wav_path}")
            else:
                print(f"íŒŒì¼ ì—†ìŒ: {mp3_path}")

    prs.save("../../data/new.pptx")
    print("ìµœì¢… PPT ìƒì„± ì™„ë£Œ: new.pptx")

# ì‹¤í–‰ ì˜ˆì œ
ppt_path = "../../data/abc.pptx"
num_slides = 10  # PPTì˜ ì´ ìŠ¬ë¼ì´ë“œ ìˆ˜
add_audio_to_ppt(ppt_path, num_slides)
```
