```python
from langchain_community.document_loaders import PyMuPDFLoader

def pdf2text(pdf_path):
    '''
    pdf_path에 있는 pdf파일을 이미지를 제외한 텍스트를 추출하여 리스트에 담는다
    리스트에 담긴 텍스트를 한글과, 영어, 숫자, 공백1칸을 제외하고 모두 파싱 후 반환
    '''
    loader = PyMuPDFLoader(pdf_path)
    docs = loader.load()

    ppts = []
    
    for doc in docs:
        # text = re.sub("[^가-힣a-zA-Z0-9\s]","",doc.page_content)
        # text = re.sub("\s+", " ", text).strip()
        # ppts.append(text)
        ppts.append(doc.page_content)
        # print(f"{text}가 append 되었슴당")
    return ppts
```

```python
pdf_path = "C:/wanted/Lang/Presentation-Agent/data/pdf/DeePrint.pdf"

full_text = pdf2text(pdf_path)
page_texts = pdf2text(pdf_path)
page_texts
```

```python
import fitz  # PyMuPDF

doc = fitz.open(pdf_path)

full_text = []
page_texts = []

for page in doc:
    text = page.get_text("text")
    if text:
        page_texts.append(text)
        full_text.append(text)  # 전체 문서용
page_images = []

# 전체 문서 기준으로 이미지 추출
for page_num in range(len(doc)):
    page = doc[page_num]
    # print(f"page : {page}")
    images = page.get_images(full=True)
    # print(f"images : {images}")

    image_data = []
    for img in images:
        # print(f"img : {img}")
        xref = img[0]
        # print(f"xref : {xref}")
        img_rects = page.get_image_rects(xref)
        # print(f"img_rects : {img_rects}")
        if img_rects:
            x0, y0, x1, y1 = img_rects[0]
            image_data.append({"image_id": xref, "x": x0, "y": y0})
    # print(f"image_data : {image_data}")
    if page_num > 5:
        break

    page_images.append({"page": page_num + 1, "images": image_data})
```

```python
# 같은 페이지에서 가장 가까운 텍스트를 이미지와 매칭
page_data = []

for page_num in range(len(doc)):
    page = doc[page_num]
    
    text_blocks = page.get_text("blocks")  # 블록별 텍스트 위치 정보
    images = page.get_images(full=True)

    page_info = {"page": page_num + 1, "texts": [], "images": []}

    for block in text_blocks:
        if len(block) >= 5:
            x0, y0, x1, y1, text = block[:5]
            page_info["texts"].append({"text": text, "x": x0, "y": y0})

    for img in images:
        xref = img[0]
        img_rects = page.get_image_rects(xref)
        if img_rects:
            x0, y0, x1, y1 = img_rects[0]
            page_info["images"].append({"image_id": xref, "x": x0, "y": y0})

    # 이미지와 가장 가까운 텍스트 매칭
    for img in page_info["images"]:
        closest_text = None
        min_distance = float("inf")

        for text_block in page_info["texts"]:
            distance = abs(text_block["y"] - img["y"])  # Y축 기준으로 가장 가까운 텍스트 찾기
            if distance < min_distance:
                min_distance = distance
                closest_text = text_block["text"]

        img["matched_text"] = closest_text if closest_text else "No text found"

    page_data.append(page_info)
```

```python
# 전체 문서에서 주요 키워드 도출
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = vectorizer.fit_transform([" ".join(full_text)])  # 전체 문서 기준
feature_names = vectorizer.get_feature_names_out()

full_keywords = sorted(zip(tfidf_matrix.toarray()[0], feature_names), reverse=True)[:15]
full_keywords = [k[1] for k in full_keywords]

print("전체 문서 키워드:", full_keywords)
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
# 전체 문서 요약

parser = PlaintextParser.from_string(" ".join(full_text), Tokenizer("english"))
summarizer = LsaSummarizer()
summary = summarizer(parser.document, 5)  # 5문장 요약

print("전체 문서 요약:")
for sentence in summary:
    print(sentence)
```

```python
page_summaries = []

for i, text in enumerate(page_texts):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    summary = summarizer(parser.document, 2)  # 페이지별 2문장 요약
    page_summaries.append({"page": i + 1, "summary": [str(sent) for sent in summary]})
```

```python
final_summary = {
    "전체 문서 키워드": full_keywords,
    "전체 문서 요약": [str(sent) for sent in summary],
    "페이지별 요약": page_summaries,
    "페이지별 이미지 매칭": page_data
}

import json
print(json.dumps(final_summary, indent=4, ensure_ascii=False))
```
